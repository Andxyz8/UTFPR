### Importando módulos necessários

```python
import requests
from bs4 import BeautifulSoup
```

### Obtendo todo o conteúdo da página

```python
url = "https://www.apontador.com.br/local/sp/campinas/faculdades_e_universidades/88Q2E372/unicamp.html."

# Obtendo o conteúdo da página por meio de uma requisição get HTTP
conteudo = requests.get(
    url, 
    verify = False
)

##transforma o conteudo da pagina em um objeto BeautifulSoup
bs4_apontador_unicamp = BeautifulSoup(conteudo.content,'html.parser')

div_avaliacoes_bruto = bs4_apontador_unicamp.find("div", {"id":"avaliacoes"})

print(div_avaliacoes_bruto)
```

### Manipulação de strings e de tags html para obter as informações das avaliações

```python
nome_avaliadores = div_avaliacoes_bruto.findAll('span', {'class': 'name'})
comentario_avaliadores = div_avaliacoes_bruto.findAll('div', {'class': 'store-comments__text'})
avaliacoes_realizada = div_avaliacoes_bruto.findAll('div', {'class': 'rate-stars'})
data_avaliacoes = div_avaliacoes_bruto.findAll('span', {'class': 'rate-stars__amount'})

for nome, comentario, avaliacao, data_avaliacao in zip(
    nome_avaliadores,
    comentario_avaliadores,
    avaliacoes_realizada,
    data_avaliacoes
):
    nome_atual = str(nome).replace('<span class="name">', '').replace('</span>', '')
    comentario_atual = str(comentario).replace(
        '<div class="store-comments__text">\n', ''
    ).replace('<p>', '').replace('</p>', '').replace('</div>', '')

    len_antes = len(str(avaliacao))
    avaliacao_atual = str(avaliacao).replace('star-checked.svg', '')
    len_depois = len(avaliacao_atual)

    qtd_estrelas_atual = (len_antes - len_depois)/16

    data_avaliacao_atual = str(data_avaliacao).replace(
        '<span class="rate-stars__amount">publicada em ', ''
    ).replace('</span>', '')

    print(f"Avaliador: {nome_atual}")
    print(f"Avaliação: {qtd_estrelas_atual}")
    print(f"Data Publicação: {data_avaliacao_atual}")
    print(f"Comentario: {comentario_atual}")
```

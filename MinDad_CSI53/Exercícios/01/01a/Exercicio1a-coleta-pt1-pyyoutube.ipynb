### Obtendo os 50 vídeos com pesquisando por "política"
- https://developers.google.com/youtube/v3/docs/search/list#parameters
- https://developers.google.com/youtube/v3/docs/search#resource
- https://developers.google.com/youtube/v3/docs/videos#statistics.commentCount

```python
busca_por_eleicao = yt_api.search(
    q = "eleição|eleicao",
    location = "-23.548887, -46.638801",
    location_radius = "30mi",
    published_after = "2022-10-02T00:00:00Z",
    published_before = "2022-10-30T00:00:00Z",
    order = "viewCount",
    count = 50,
    safe_search = "moderate",
    search_type= "video",
    return_json = True
)
busca_por_eleicao
```

### Visualizando informações retornadas

```python
video_items = busca_por_eleicao["items"]
video_items
```

### Obtendo informações dos 50 vídeos com mais visualizações de "política" em São Paulo - SP na época do segundo turno das eleições presidências
- https://developers.google.com/youtube/v3/docs/videos#statistics.commentCount

```python
prop_videos = []

for video in video_items:
    print(f"Título: {video['snippet']['title']}")
    print(f"\tvideoId: {video['id']['videoId']}\n")

    video_props = yt_api.get_video_by_id(
        video_id = video['id']['videoId'],
        return_json = True
    )

    prop_videos.append(video_props)

    # Esperar 250ms por requisição por desencargo de consciência
    sleep(0.25)
```

### Visualizando a lista que foi retornada

```python
prop_videos
```

### Ordernando lista de dicionarios por comentarios

```python
prop_videos_sorted = sorted(
    prop_videos, # Lista de dicionários
    key = lambda x: int(x["items"][0]["statistics"]["commentCount"]), # Queremos ordenar por número de comentários, lambda para acessar o valor do dicionário
    reverse = True # Queremos ordem decrescente pro item[0] ser aquele com mais comentários	
)
prop_videos_sorted
```

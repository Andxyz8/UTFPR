### Obtendo os 50 v√≠deos com pesquisando por "pol√≠tica"
- https://developers.google.com/youtube/v3/docs/search/list#parameters
- https://developers.google.com/youtube/v3/docs/search#resource
- https://developers.google.com/youtube/v3/docs/videos#statistics.commentCount

```python
busca_por_eleicao = yt_api.search(
    q = "elei√ß√£o|eleicao",
    location = "-23.548887, -46.638801",
    location_radius = "30mi",
    published_after = "2022-10-02T00:00:00Z",
    published_before = "2022-10-30T00:00:00Z",
    order = "viewCount",
    count = 50,
    safe_search = "moderate",
    search_type= "video",
    return_json = True
)
busca_por_eleicao
```

### Visualizando informa√ß√µes retornadas

```python
video_items = busca_por_eleicao["items"]
video_items
```

### Obtendo informa√ß√µes dos 50 v√≠deos com mais visualiza√ß√µes de "pol√≠tica" em S√£o Paulo - SP na √©poca do segundo turno das elei√ß√µes presid√™ncias
- https://developers.google.com/youtube/v3/docs/videos#statistics.commentCount

```python
prop_videos = []

for video in video_items:
    print(f"T√≠tulo: {video['snippet']['title']}")
    print(f"\tvideoId: {video['id']['videoId']}\n")

    video_props = yt_api.get_video_by_id(
        video_id = video['id']['videoId'],
        return_json = True
    )

    prop_videos.append(video_props)

    # Esperar 250ms por requisi√ß√£o por desencargo de consci√™ncia
    sleep(0.25)
```

### Visualizando a lista que foi retornada

```python
prop_videos
```

### Ordernando lista de dicionarios por comentarios

```python
prop_videos_sorted = sorted(
    prop_videos, # Lista de dicion√°rios
    key = lambda x: int(x["items"][0]["statistics"]["commentCount"]), # Queremos ordenar por n√∫mero de coment√°rios, lambda para acessar o valor do dicion√°rio
    reverse = True # Queremos ordem decrescente pro item[0] ser aquele com mais coment√°rios	
)
prop_videos_sorted
```
